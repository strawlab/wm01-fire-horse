````python
# You must run this cell, but you can ignore its contents.
import hashlib
import matplotlib.pyplot as plt

def ads_hash(ty):
    """Return a unique string for input"""
    ty_str = str(ty).encode()
    m = hashlib.sha256()
    m.update(ty_str)
    return m.hexdigest()[:10]
````
