# Slicing

## list slicing

```python
x=["red","orange","yellow","green","blue","indigo","violet"]
x[0:3]
```

```python
x[3]
```

```python
x[:3]
```

```python
x=["red","orange","yellow","green","blue","indigo","violet"]
x[None:3]
```

```python
x[0:3]
```

```python
x=["red","orange","yellow","green","blue","indigo","violet"]
x[3:]
```

```python
x=["red","orange","yellow","green","blue","indigo","violet"]
x[3:None]
```

```python
x=["red","orange","yellow","green","blue","indigo","violet"]
x[3:-1]
```

## tuple slicing

```python
x=("red","orange","yellow","green","blue","indigo","violet")
x[0:3]
```

```python
x[:3]
```

## implicit coercion

```python
if 10:
    print("10 is an integer, not a boolean. Why is this not an error?")
```

```python
if 0:
    print("why doesn't this print?")
```

```python
x = "False"
if x:
    print("hello")
```

```python
x = ""
if x:
    print("hello")
```
