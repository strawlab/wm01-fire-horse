# Conditionals and loops

So far our programs have run straight through, top to bottom. **Control flow**
lets a program make decisions and repeat work. We start with the `bool` type
and comparisons, then `if`/`elif`/`else`, `assert` for checking our
assumptions, and finally `while` loops.

# Boolean (`bool`) type

Python's `bool` type can take one of two values: `True` or `False`. It is used to test a condition, such as in an *if statement*.

```python
x = 1
y = x > 0
y
```

```python
type(y)
```

```python
x = 1

if x > 0:
    print("x is positive")
    x = 10

x
```

```python
x = -1

if x > 0:
    y = 20
    print("x is positive")
    print("x is still positive")
else:
    y = 40
    print("x is negative")
    print("x is still negative")
    
print("out of the blocks")
print(y)
```

## Nested blocks and `elif`

```python
if True:
    print("statement 1")
    print("statement 2")
    print("statement 3")
```

```python
a = 1
b = -2

if a==1:
    if b>0:
        print("a is one and b is positive")
    else:
        print("here")
        print("a is one")
else:
    print("a is not one")
```

```python
a = 1
b = -1.0

if a==1:
    if b>0:
        print("a is one and b is positive")
    elif b<0:
        print("a is one and b is negative")
    else:
        print("a is one")
        print("b is zero")
else:
    print("a is not one")
```

## Checking your assumptions with `assert`

`assert` checks that a condition is `True`. If it is, nothing happens; if it is
`False`, Python raises an `AssertionError`. This is a cheap, powerful way to
catch mistakes early and to document what your code expects. We will use
`assert` throughout the course to test small examples.

```python
assert True
```

```python
assert False
```

```python
assert 2 + 2 == 4
```

```python
assert 1 + 1 == 3
```

```python
assert 1 + 1 == 3, "This is the error message"
```

```python
def my_cool_function(a,b):
    assert type(a)==list
    assert type(b)==list
    a.append(10)
    b.append(20)
```

```python
x = [1,2,3]
y = [4,4,5]
my_cool_function( x, y)
```

```python
x
```

```python
y
```

```python
my_cool_function(1,2)
```

# Control flow with `while`

```python
x = True
y = 0
while x:
    print(y)
    y = y + 100
    if y >= 1000:
        x = False
print('done')
```

```python
y = 0
while True:
    print(y)
    y = y + 100
    if y >= 1000:
        break
print('done')
```

```python
y = 0
while y < 1000:
    print(y)
    y = y + 100
print('done')
```

## `break` and `continue`

```python
y = 0
while True:
    print(y)
    y = y + 100
    if y >= 1000:
        break
```

```python
y = 0
while True:
    print(y)    
    y = y + 100
    if y >= 1000:
        break
    if y < 400:
        continue    
    y = y + 10
```
