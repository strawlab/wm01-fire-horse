# Strings and f-strings

We often need to build strings from values — a label for a plot, a message for
a user, a line to write to a file. Python has several ways to do this. Today's
recommended tool is the **f-string**, but you will see the older `%` and
`.format()` styles in existing code, so we show them briefly first.

## Old-style string formatting with `%`

When the operator `%` is used on a string, the string is used as a *format string* for old-style formatting.

In these old-style format strings, `%d` means to print an integer, `%s` means to print a string.

```python
"The numbers are %d, %d, %d" % (5,10,20)
```

```python
tmp = (5,10,20)
"The numbers are %d, %d, %d" % tmp
```

```python
tmp = (5,10,20,100)
"The numbers are %d, %d, %d" % tmp
```

```python
my_string = "The numbers are %d, %d, %d"
my_string
```

```python
my_string % (7, 14, 21)
```

## New-style (since Python 2.6) formatting with `.format()`

```python
"The numbers are {}, {}, {}".format(5,10,20)
```

```python
"Hello {}".format("world")
```

## Even newer (since Python 3.6) style formatting with f-strings.

```python
name="Andrew"
my_string = f"Hello, my name is {name}."
print(my_string)
```

```python
f"Hello, my name is {name}."
```

```python
f"Hello, my name is {name}. {akjfhasd}"
```

```python
f"Hello, my name is {name}. {{akjfhasd}}"
```

```python
"Hello, my name is {}.".format(name)
```

```python
my_template = "hello {}"
print('my template is:', my_template)
my_template.format(name)
```

```python
name = "different name"
my_template.format(name)
```

## Strings are sequences

A string is a *sequence* of characters, so — just like a list — you can index
it, slice it, and iterate over it with a `for` loop. Each item you get from
iterating is a one-character string. (We rely on this in the exercise.)

```python
dna = "GATTACA"
print(dna[0])      # indexing
print(dna[1:4])    # slicing
for base in dna:   # iterating
    print(base)
```
