# Plotting with matplotlib

[matplotlib](https://matplotlib.org/) is the most widely used plotting library in
scientific Python. Almost every figure you see in a Python-based paper was made
either with matplotlib directly or with a library (like `seaborn`, which we use
later today) built on top of it.

We will use just one part of matplotlib, called `pyplot`. By long-standing
convention it is imported and abbreviated as `plt`:

````python
import matplotlib.pyplot as plt
````

In a Jupyter notebook, matplotlib figures appear automatically underneath the
cell that creates them. (Further down we show how to display or save a figure when
you run a plain Python program *outside* Jupyter.)

## A first line plot

The workhorse function is `plt.plot()`. If you give it a single list of numbers,
it plots them on the Y axis against their position (0, 1, 2, ...) on the X axis.

Let's make some data to plot. Recall from Lecture 2 how a population grows when
each generation is `Rt` times the size of the previous one. Here is a small `for`
loop that builds a list of population sizes:

````python
def growth_curve(Rt, n_generations=15):
    number_infected = 1.0
    values = []
    for generation in range(n_generations):
        values.append(number_infected)
        number_infected = number_infected * Rt
    return values

growth_curve(1.5)
````

Now we plot it:

````python
plt.plot(growth_curve(1.5));
````

Notice the semicolon (`;`) at the end of the line. `plt.plot()` returns a value
(a list of the line objects it drew), and in Jupyter that value would be printed
above the figure. The trailing semicolon just suppresses that text output. It is
optional, but it keeps the display tidy.

## Labelling the axes

A plot without axis labels is almost useless in science: nobody can tell what the
numbers mean. Always add labels with `plt.xlabel()` and `plt.ylabel()`, and a
title with `plt.title()`. All of these take a string.

````python
plt.plot(growth_curve(1.5))
plt.xlabel("generation")
plt.ylabel("number infected")
plt.title("Exponential growth");
````

## Plotting X against Y, and multiple lines

Usually you have both an X and a Y value for each point. Then you call
`plt.plot(x, y)`. You can also call `plt.plot()` several times before the figure is
shown to draw more than one line on the same axes.

To tell the lines apart, pass a `label=` to each `plt.plot()` call and then call
`plt.legend()` once. The text between dollar signs (e.g. `"$R_t$"`) is rendered as
a mathematical symbol.

````python
generations = list(range(15))

plt.plot(generations, growth_curve(1.1), label="$R_t$ = 1.1")
plt.plot(generations, growth_curve(1.5), label="$R_t$ = 1.5")
plt.plot(generations, growth_curve(2.0), label="$R_t$ = 2.0")
plt.xlabel("generation")
plt.ylabel("number infected")
plt.legend();
````

## Scatter plots

To plot individual data points *without* a connecting line, pass a *format string*
as the third argument. `'o'` means "draw a circle at each point and do not connect
them". Other useful format characters include `'.'` (small dots), `'s'` (squares)
and `'x'` (crosses).

Let's load some real data. This is the Iris dataset, which we read here with plain
Python exactly as in Lecture 2 (later today we will do the same thing far more
easily with pandas). The result is a dictionary mapping each column name to a list
of values.

````python
fobj = open("iris.csv")
iris = {}
for line_num, line in enumerate(fobj.readlines()):
    entries = line.strip().split(',')
    if line_num == 0:
        column_names = entries
        for column_name in column_names:
            iris[column_name] = []
        continue
    for (column_name, entry) in zip(column_names, entries):
        if column_name != 'species':
            entry = float(entry)
        iris[column_name].append(entry)

list(iris.keys())
````

Now we make a scatter plot of two of the measurements against each other:

````python
plt.plot(iris['sepal_width'], iris['petal_width'], 'o')
plt.xlabel("sepal width [cm]")
plt.ylabel("petal width [cm]");
````

### Overlapping points and transparency

Many points fall on top of each other, so it is hard to see where the data is
dense. Setting `alpha` (a number between 0 and 1) makes each marker partly
transparent, so overlapping points show up as darker regions.

````python
plt.plot(iris['sepal_width'], iris['petal_width'], 'o', alpha=0.2)
plt.xlabel("sepal width [cm]")
plt.ylabel("petal width [cm]");
````

## Histograms

A histogram shows the *distribution* of a single set of numbers by counting how
many values fall into each of a series of bins. Use `plt.hist()`. The `bins`
argument controls how many bins are used.

````python
plt.hist(iris['petal_length'], bins=20)
plt.xlabel("petal length [cm]")
plt.ylabel("count");
````

Notice the two clumps: this hints that the dataset contains more than one
species with different petal lengths.

## Using matplotlib outside Jupyter

In Jupyter the figure is shown for you. When you write a **plain Python program**
(a `.py` file that you run from the command line) nothing is displayed unless you
ask for it. You have two choices:

Open an interactive window on your screen:

```python
plt.show()
```

Save the figure to an image file (e.g. a PNG), which is what you usually want in a
script that produces results:

```python
plt.savefig("my_plot.png")
```

You will use `plt.savefig()` in today's exercise, where you write a standalone
program that reads a CSV file and saves a plot of it.
