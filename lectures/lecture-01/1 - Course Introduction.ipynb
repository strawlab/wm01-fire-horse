## Computational thinking

Before we write any code, it's worth naming the mental habits that let us
solve problems with a computer. These transfer far beyond Python, and we will
come back to them again and again in this course:

- **Decomposition** — breaking a big, intimidating problem into smaller
  pieces you can tackle one at a time.
- **Abstraction** — ignoring irrelevant detail so you can focus on the
  structure that matters (a function, a variable with a good name, a
  simplified model of a system).
- **Pattern recognition** — noticing that a new problem is *like* one
  you've already solved, and reusing the approach.

**Discussion:** work through the bee-navigation example above using these
three ideas. How would you **decompose** "how does a bee find its way home?"
into smaller sub-questions? What could you **abstract away** to make the
problem tractable — e.g. treating the bee's path as a sequence of headings
and distances rather than raw sensor data? What **patterns** from other
navigation problems, such as GPS or dead reckoning, might apply?

We'll notice these three ideas again in a few minutes when we deconstruct a
Blockly program.

## The AI-assisted landscape

Most professional code today is written with the help of AI assistants —
chat-based tools and increasingly *agentic* coding tools that can read, run,
and edit code on their own. This raises an obvious question: why learn the
fundamentals below at all?

Because using these tools well — writing a good prompt, recognizing a
plausible-but-wrong answer, verifying that generated code actually does what
you asked — requires the same computational thinking this course teaches. We
will use AI assistants throughout the course (starting from Lecture 4), but
always alongside, not instead of, understanding what the code does.
