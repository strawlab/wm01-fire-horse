# Blockly deconstructed

Consider this [Blockly program](https://blockly.games/bird?lang=en&level=2#562ubo).

![blockly-bird](blockly-bird.png)

This is actually a full computer program! When you successfully complete it, you can see the automatic translation to JavaScript. JavaScript is a different programming language from Python, but the important programming ideas transfer directly.

```javascript
if (noWorm()) {
  heading(0);
} else {
  heading(90);
}
```

If we rewrote this in Python, it would be:

```python
if noWorm():
    heading(0)
else:
    heading(90)
```

Let's discuss what is going on here.

A program is a precise set of instructions executed by a computer. In this Blockly example, the program is small, but it already has most of the ingredients we will see all week:

- **Input from the world**: `noWorm()` asks a question about the game state.
- **A decision**: `if ... else ...` chooses between two possible actions.
- **Output to the world**: `heading(0)` or `heading(90)` changes what the bird does.
- **Order**: the computer checks the condition before it chooses a heading.
- **Names**: `noWorm` and `heading` are names for behavior supplied by the Blockly game.

The operating system, browser, Blockly runtime, and game engine hide many details from us. That is useful: we can think about the logic of the bird without thinking about graphics hardware, keyboard events, or JavaScript internals.

## Computational thinking in the bird program

This tiny program is a useful example of the three computational-thinking ideas from the course introduction.

**Decomposition**: split the problem "get the bird to the worm" into smaller questions: Is the worm above or below? Which heading moves the bird toward it? What should happen in each case?

**Abstraction**: ignore the full visual scene and represent the important part as a yes/no question, `noWorm()`. We do not need to know every pixel in the game window to decide what to do next.

**Pattern recognition**: this has the same shape as many biological and data-analysis problems: measure something, choose between alternatives, act or classify. For example: if a measurement is above a threshold, label a cell as active; otherwise label it inactive.

A function lets us give a name to a small piece of behavior and reuse it. In Blockly, the function is represented as a custom block. In Python, we will write the same idea with `def`:

```python
def play_phrase():
    # steps go here
    pass

play_phrase()
```

The definition says what the function means. The call runs it. Keeping those two ideas separate is essential: writing down a recipe is not the same as cooking the meal.
