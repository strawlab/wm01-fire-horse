## Functions

A building block of programming are **functions**. These are pieces of code that are called (also known as executed) with input and return output. They may also have side effects.

Let's look at a function definition:

```python
def my_function_name(argument1, argument2):
    variable1 = argument1 + argument2
    print("hello from my function") # this is a side effect
    return variable1
```

Above, we created a function named `my_function_name`. This function takes two inputs, called arguments. There are several steps performed in the function. First, we add `argument1` to `argument2` and store the result in `variable1`. Then we print something. Then we return `variable1` as the output of our function.

The indentation (spaces) at the beginning of the lines in the function tell Python which lines of code are part of the function. Everything with the same level of indentation is a block.

Above, we called the built-in function `print()`.

The `print()` function takes any number of arguments as input, returns `None`, and as a side effect prints the arguments to the screen.

````python
def triple(x):
    return x * 3
````

````python
triple(4)
````

A function can have more than one argument.

````python
def weighted_sum(a, b):
    return 5 * a + 3 * b
````

````python
weighted_sum(3, 10)
````

A function can also call another function. Predict the result before running this cell.

````python
def triple_thrice(x):
    y = triple(x)
    z = triple(y)
    return triple(z)

triple_thrice(2)
````

Variables created inside a function live in the function's local scope. They do not replace variables with the same name outside the function.

````python
x = 43

def subtract_inside_function(x):
    x = x - 26
    return x

inside_value = subtract_inside_function(2)
print(inside_value)
print(x)
````

A common beginner mistake is to print a value when the task asks a function to return a value.

````python
def prints_but_returns_none(x):
    print(x * 30)

result = prints_but_returns_none(10)
print(result)
````

## The predict → run → inspect → debug cycle

Programming is rarely about writing correct code on the first try. Instead,
it's an iterative loop, and it's worth naming explicitly because you'll repeat
it constantly for the rest of this course (and your career):

1. **Predict** what a piece of code will do before you run it — try this with
   the Python Tutor above, guessing the value of a variable before clicking
   "Next".
2. **Run** it.
3. **Inspect** what actually happened — the printed output, the variable
   values, or an error message.
4. **Debug** — reconcile any difference between what you predicted and what
   happened, then repeat.

The three cells below are a good place to practice: before running each one,
